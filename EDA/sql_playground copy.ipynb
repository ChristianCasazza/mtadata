```python
import bauplan
```

```python
sql_query = """
SELECT 
    pickup_datetime,
    dropoff_datetime,
    PULocationID,
    DOLocationID,
    trip_miles,
    trip_time,
    base_passenger_fare,
    tolls,
    sales_tax,
    tips
FROM 
    taxi_fhvhv
WHERE 
    pickup_datetime >= '2022-12-01T00:00:00-05:00' 
    AND pickup_datetime < '2023-01-01T00:00:00-05:00'

LIMIT 20
"""
```

```python
# instantiate the sdk client 
client = bauplan.Client()
# pass the branch and the sql expression to the method query
current_branch = "main"

# run a query and get in return an arrow table
table = client.query(sql_query, max_rows=None, branch_name=current_branch)
# convert the arrow table into a Pandas DataFrame
data = table.to_pandas()
# display the Pandas DataFrame
data
```

```python
import pandas as pd
import math



df = data
# create time filter on datetime UTC
time_filter = pd.to_datetime('2022-01-01')
time_filter_utc = time_filter.tz_localize('UTC')
# filter df by timestamp
df = df[df['pickup_datetime'] >= time_filter_utc]
# exclude rows with trip_miles = 0
df = df[df['trip_miles'] > 0.0]
# exclude rows with trip_miles > 200
df = df[df['trip_miles'] < 200.0]

# return a Pandas dataframe
df
```

```python
import duckdb
sql_query = f"""
    SELECT * from df
    """
# run the query and return the results as an Arrow Table
new = duckdb.sql(sql_query).arrow()

display = table.to_pandas()
display
```

```python
import duckdb
sql_query = f"""
    SELECT 
        COUNT(df.pickup_datetime) AS number_of_trips, 
        df.PULocationID,
        taxi_zones.Borough, 
        taxi_zones.Zone
    FROM df
    JOIN taxi_zones 
    ON df.PULocationID = taxi_zones.LocationID
    GROUP BY 
        df.PULocationID, 
        taxi_zones.Borough, 
        taxi_zones.Zone
    ORDER BY 
        number_of_trips DESC;
    """
# run the query and return the results as an Arrow Table
new = duckdb.sql(sql_query).arrow()


new
```
